# Pandas GroupBy Function

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('E:\\docs and imp.videos\\AI video in hindi from yt\\machine learning\\dataset\\STUDENT_10_without_date.csv')
df
```

```python
gr1 = df.groupby(by = 'section')
gr1
```

```python
gr1.groups
```

```python
gr2 = df.groupby(by = 'Class')
gr2
```

```python
gr2.groups
```

```python
df.groupby(['section','Class']).groups
```

```python
for section,df_1 in gr1:
    print(section)
    print(df_1)
```
