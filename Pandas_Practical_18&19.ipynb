# Pandas Concat #1

```python
import pandas as pd
import numpy as np
```

```python
sr1 = pd.Series([0,1,2,3,4])
sr1
```

```python
sr2 = pd.Series([5,6,7])
sr2
```

```python
pd.concat([sr1,sr2])
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
             'Name':['A','B','C','D'],
             "Class":[9,10,11,12]})
df1
```

```python
df2 = pd.DataFrame({'ID':[9,10,11,12],
             'Name':['E','F','G','H'],
             "Class":[5,6,7,8]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df1],axis = 1)
```

```python
pd.concat([df1,df2],axis = 1)
```

```python
pd.concat([df1,df2],axis = 0)
```

```python
pd.concat([df1,df2],axis = 0,ignore_index = True)
```

# Pandas Concat #2

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
             'Name':['A','B','C','D'],
             "Class":[9,10,11,12]})
df1
```

```python
df2 = pd.DataFrame({'ID':[1,2,],
             'Name':['A','B'],
             "Class":[11,12]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis = 1)
```

```python
pd.concat([df1,df2],axis = 1,join = 'inner')
```

```python
pd.concat([df1,df2],axis = 1, join = 'outer')
```

```python
df1
```

```python
df2 = pd.DataFrame({'ID':[9,10,11,12],
             'Name':['E','F','G','H'],
             "Class":[5,6,7,8]})
df2
```

```python
pd.concat([df1,df2],keys=["df1","df2"])
```

```python
pd.concat([df1,df2],keys=["First DataFrame","Second DataFrame"])
```

```python
pd.concat([df1,df2],keys=["First DataFrame","Second DataFrame"],axis = 1)
```

```python
pd.concat([df1,df2],keys=["First DataFrame","Second DataFrame"],join='outer',axis = 1)
```

```python
df1
```

```python
df2 = pd.DataFrame({'marks':[20,30,23,43]})
```

```python
df2
```

```python
pd.concat([df1,df2],axis=1)
```
