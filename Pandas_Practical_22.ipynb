# Pandas Pivot Table

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('E:\docs and imp.videos\AI video in hindi from yt\machine learning\dataset\movie_Rating.csv')
```

```python
df
```

```python
df.pivot_table(index = 'Genre')
```

```python
df.pivot_table(index = 'Genre',columns = 'Year')
```

```python
df.pivot_table(index = 'Genre',columns = 'Year', aggfunc = 'count')
```

```python
df.pivot_table(index = 'Genre',columns = 'Year', aggfunc = 'max')
```

```python
df.pivot_table(index = 'Genre',columns = 'Year', aggfunc = 'min')
```

```python
df.pivot_table(index = 'Genre',columns = 'Year', aggfunc = 'mean')
```

```python
df.pivot_table(index = 'Genre',columns = 'Year', aggfunc = 'sum')
```

```python
df.pivot_table(index = 'Genre',columns = 'Year', aggfunc = 'max',fill_value = '1')
```

```python
df.pivot_table(index = 'Genre',columns = 'Year', aggfunc = 'max',fill_value = 'none')
```

```python
df.pivot_table(index = 'Genre',columns = 'Year',fill_value = '0')
```

```python
df.pivot_table(index = 'Genre',columns = 'Year',margins = True,fill_value = '0')
```

```python
df.pivot_table(index = 'Genre',columns = 'Year',margins = True,fill_value = '0',aggfunc = 'sum')
```

```python
df.pivot_table(index = 'Genre',columns = 'Year',margins = True,fill_value = '0',aggfunc = 'min')
```
